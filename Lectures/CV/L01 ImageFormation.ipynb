```python
print(len(img))
out=np.zeros(img.shape[0:2])
for x in range(0,len(img)):
    for y in range(0,len(img[0])):
        p0=img[x][y]
        out[x][y]=np.mean(p0)


    # plt.imshow(out)
    
```
