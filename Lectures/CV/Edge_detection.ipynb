# Sobel Edge Detection Filters
an example of gradient filter

```python
# Sobel Edge Detection Filters

sobel_filter=np.array([1,0,-1,2,0,-2,1,0,-1]).reshape(3,3)
sobel_filter, sobel_filter.T
```

```python
# prewitt

prewitt_filter=np.array([-1,-1,-1,0,0,0,1,1,1]).reshape(3,3).T
prewitt_filter, prewitt_filter.T
```

```python
# to find difference in diagonal element
# roberts
roberts_filter=np.array([0,1,-1,0]).reshape(2,2)
roberts_filter,roberts_filter.T
```
