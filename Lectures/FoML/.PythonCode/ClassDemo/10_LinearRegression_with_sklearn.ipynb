# Linear Regression in python

```python
import numpy as np
from random import random
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Generate random points for linear regression
$$y=mx+c$$
assume $m$ and $c$ add some $noise$

```python

x=np.arange(0,10)
m=-2#slope
noise=10
c=15#c constant
noiseDrift=[round(2*noise*random()-noise,2) for n in range(0,len(x))]
y=[(y*m+c)+noiseDrift[i] for i,y in enumerate(x) ]
y=np.array(y)
fig,ax=plt.subplots(1,figsize=(3,3))
ax.plot(hold=True)
ax.grid()
ax.scatter(x,y)
```

# Regression with sklearn
### model and train

```python
model=LinearRegression()

X=x.reshape(-1,1)
Y=y.reshape(-1,1)

model.fit(X,Y)
model.score(X,Y),model.coef_,model.intercept_
```

### model predict

```python
xtest=np.arange(0,10,0.1)
xt=xtest.reshape(-1,1)
yp=model.predict(xt)
```

# Estimate yPredict with m x and c

## Validate with plot

```python
def testPlot(x,y,xt,yt):
    fig,ax=plt.subplots(1,figsize=(8,8))
    ax.plot(hold=True)
    ax.grid()
    ax.scatter(x,y,label='Actual data points')

    ax.plot(xt,yt,'-r',label='Estimated line with linear regression')
    ax.legend()

testPlot(X,Y,xt,yp)
    
```
