```python
w,v=np.linalg.eig(A)
w,v

v1=v[0,:]
v2=v[1,:]
o=np.array([0,0])
v,v1,v2,o
```

```python
# plt.figure()
fig,ax=plt.subplots(2,2,figsize=(8,8))
ax[0,0].grid()
ax[0,0].scatter(A[:,0],A[:,1],color='red')



ax[0,1].axis([-2,2,-2,2])
ax[0,1].scatter(A[:,0],A[:,1],color='red')
ax[0,1].grid()
ax[0,1].plot([o,v1],color='green',label='Eigen Vector 1')
ax[0,1].plot([o,v2],color='blue',label='Eigen Vector 2')

ax[1,1].plot([o,v1],color='green',label='Eigen Vector 1')

```
