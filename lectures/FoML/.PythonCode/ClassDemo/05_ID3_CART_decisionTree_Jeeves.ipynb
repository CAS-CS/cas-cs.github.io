```python
Humidity="H,H".split(",")  
Windy='T,F'.split(",") 
PlayGolf="Y,N".split(",")
df=pd.DataFrame()
df['Humidity']=Humidity
df['Windy'] =Windy
df['PlayGolf']=PlayGolf
data=df
```

```python
class Entropy:
    def __init__(self,df):
        self.df=df
        self.H=self.EntropyCalc()
    
    def EntropyCalc(self):
        df=self.df
        self.S=list(df.iloc[:, -1].values)
        self.set=list(set(self.S))
        p=[]
        count=[]
        for i,val in enumerate(self.set):
            count.append(self.S.count(val))
            p.append(self.S.count(val)/len(self.S))
        self.count=count
        self.P=np.array(p)
        self.H=np.sum(-1*self.P*np.log2(self.P))
        return self.H

    def __repr__(self):
        print("Input Data Frame")
        print(f"{self.df}\n")

        print(f"There are {len(self.S)} elements in sample space")
        print(f"Sample space: \n{self.S}")
        print(f"Unique values:\t{self.set}")
        print(f"Count of Unique values:\t {self.count}")
        print(f"Probability:\t {self.P}")
        print(f"\nEntropy={self.H}")
        return ''

print(Entropy(data))
main=Entropy(data)
H=main.H
H
```

```python
def Prob(S,event):
    # print(S,event)
    S=list(S)
    event=event

    # print(S,event)
    p=[]
    p.append(S.count(event)/len(S))
    # print(p)
    return p

# S=list(data["OUTLOOK"].values)
# Prob(S,'Sunny')
```

```python
Hmain=Entropy(data).H
InformationGain=[]
print(f"Entropy of Complete data set i {Hmain}")
S=list(data.head(0))
for i,col in enumerate(S):
    # print(i,len(list(data.head(0)))-1)
    if i==len(list(data.head(0)))-1:
        break
    elements=list(data[col])
    uElement=set(elements)
    print('\n\n\n=========================')
    print(f"Testing for {col}")
    print('=========================\n')
    print(f"Entries are: {elements}")
    print(f"Unique entries are: {uElement}")
    
    print('\n')
    Imat=[]
    for e in uElement:
        
        print('\n-------------------------------------')
        print(f"Checking for {col} == {e}")
        print('-------------------------------------\n')
        df=data[data[col]==e]
        P=Prob(data[col],e)[0]
        H=Entropy(df).H
        I=P*H
        print(P,H,I)
        Imat.append(I)
        print(f"Probability of {col} == {e} is {P}")
        df=df.drop(col,axis=1)
        print(f"Entropy of {col} == {e} is {H}")
        print(f"Information Gain of {col} when {e} is {I}")
        
        
        print('\n')
        print(df)
        print('\n')

    IG=Hmain-np.sum(Imat)
    InformationGain.append(IG)
    print("\n\n\n++++++++++++++++++++++++++++++++++")
    print(f'Information gain of {col} with is {IG}')
    print("++++++++++++++++++++++++++++++++++\n\n")

print("\n\n\n********************************************")
print(f'Information gain MATRIX {S[0:-2]} with is {InformationGain}')
print("********************************************")

```
