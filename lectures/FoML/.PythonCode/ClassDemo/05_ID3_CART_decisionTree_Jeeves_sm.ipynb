```python
class Entropy:
    def __init__(self,df):
        self.df=df
        self.H=self.EntropyCalc()
    
    def EntropyCalc(self):
        df=self.df
        self.S=list(df.iloc[:, -1].values)
        self.set=list(set(self.S))
        p=[]
        count=[]
        for i,val in enumerate(self.set):
            count.append(self.S.count(val))
            p.append(self.S.count(val)/len(self.S))
        self.count=count
        self.P=np.array(p)
        self.H=np.sum(-1*self.P*np.log2(self.P))
        return self.H

    def __repr__(self):
        print("Input Data Frame")
        print(f"{self.df}\n")

        print(f"There are {len(self.S)} elements in sample space")
        print(f"Sample space: \n{self.S}")
        print(f"Unique values:\t{self.set}")
        print(f"Count of Unique values:\t {self.count}")
        print(f"Probability:\t {self.P}")
        print(f"\nEntropy={self.H}")
        return ''

print(Entropy(data))
main=Entropy(data)
H=main.H
H
```

```python
class GenerateTree:
    def __init__(self,data):
        self.data=data
        H= self.Entropy(data)
        self.H=H
        if H==0:
            Tree=self.SelectRootNode(data)

            self.tree='false'
            
        else:
            Tree=self.SelectRootNode(data)
            # self.tree=Tree
            self.type='decisionNode'
            self.name=Tree['ParentNode']
            # print(Tree)
        # print(self)

    def __repr__(self):
        T=self.tree
        print(f"\nDataFrame\n")
        print(f"{self.data}")
        print(f"Entropy of data = {self.H}")
        print(f"Type of node is = {self.type}")
        if T!='false':#not a leafNode
            print(f"Max information gain with = {self.name}\n")
            for i,t in enumerate(T['child']):
                print(f"\n{self.name}={t['node']} Sub data")
                print(t['data'])
                print(GenerateTree(t['data']))
        else:#leafNode
            print(f"Decision = {self.name}\n")
                
        return ""
        

    
    @staticmethod
    def Entropy(df):
        S=list(df.iloc[:, -1].values)
        Sunique=list(set(S))

        p=[]
        count=[]
        for i,val in enumerate(Sunique):
            count.append(S.count(val))
            p.append(S.count(val)/len(S))
        count=count
        
        P=np.array(p)
        H=np.sum(-1*P*np.log2(P))
        return H
    
    @staticmethod
    def Probability(S,event):
        S=list(S)
        p=[]
        p.append(S.count(event)/len(S))
        return p
    
    def SelectRootNode(self,data):
        Hmain=self.Entropy(data)
        if Hmain==0:
            self.type='leafNode'
            self.name=data.iloc[-1,-1]
            tree=dict(ParentNode='False',child='none')
            return tree
        InformationGain=[]
        S=list(data.head(0))
        for i,col in enumerate(S):
            if i==len(list(data.head(0))):
                break
            elements=list(data[col])
            uElement=set(elements)
            for e in uElement:
                df=data[data[col]==e]
                P=self.Probability(data[col],e)[0]
                H=self.Entropy(df)
                I=P*H
                Imat.append(I)

                df=df.drop(col,axis=1)


            IG=Hmain-np.sum(Imat)

            InformationGain.append(IG)
        ParentNode='none'
        ParentNode=S[np.argmax(InformationGain)]

        ChildNode=list(set(data[ParentNode].values))
        
        childData=[]
        child=[]
        for i,c in enumerate(ChildNode):
            subdf=data[data[ParentNode]==c]
            subdf=subdf.drop(ParentNode,axis=1)
            subT=GenerateTree(subdf)
            # print(subT)
            child.append(dict(node=c,data=subdf,tree=dict(subtree=subT)))


        tree=dict(ParentNode=ParentNode,child=child)
    
    

        self.tree=tree
        
        return tree


      
T=GenerateTree(data)
print(T)
```
