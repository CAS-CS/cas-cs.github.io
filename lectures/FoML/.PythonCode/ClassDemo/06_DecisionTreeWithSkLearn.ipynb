```python
from sklearn import tree
X = [[0, 0], [1, 1]]
Y=[0,1]
df=pd.DataFrame(dict(X=X,Y=Y))
print(df)
clf=tree.DecisionTreeClassifier()
clf=clf.fit(X,Y)
clf.predict([[0.,1.]])
```

```python
clf.predict_proba([[1., 2.]])
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```
